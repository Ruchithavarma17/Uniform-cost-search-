```python
import heapq

def ucs(graph, start, goal):
    # Priority Queue: (cost, node, path)
    pq = [(0, start, [start])]
    explored = set()
    step = 1

    while pq:
        print(f"\nSTEP {step}")

        print("Priority Queue (before pop):")
        for item in pq:
            print(item)

        cost, node, path = heapq.heappop(pq)

        print(f"\nNode selected : {node}")
        print(f"Cost so far : {cost}")
        print(f"Path : {path}")

        if node == goal:
            print("\nGoal reached!")
            return cost, path

        if node in explored:
            continue

        explored.add(node)
        print("Explored set :", explored)

        for neighbor, edge_cost in graph.get(node, []):
            new_cost = cost + edge_cost
            heapq.heappush(pq, (new_cost, neighbor, path + [neighbor]))
            print(f"Added to pq : {(new_cost, neighbor, path + [neighbor])}")

        step += 1

    return None, None


graph = {
    'A': [('B', 2), ('C', 4)],
    'B': [('D', 3), ('E', 4)],
    'C': [('F', 5)],
    'D': [('G', 2)],
    'E': [('H', 3)],
    'F': [('I', 4), ('I', 6)],
    'G': [('J', 2)],
    'H': [('K', 3)],
    'I': [('K', 8)],
    'J': [('L', 2)],
    'K': [('M', 3)],
    'L': [],
    'M': [('N', 4)],
    'N': []
}

start = 'A'
goal = 'N'

cost, path = ucs(graph, start, goal)

print("\nFINAL RESULT")
print("Optimal cost :", cost)
print("Optimal path :", " -> ".join(path))
```
